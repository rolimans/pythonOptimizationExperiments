```python
import pandas as pd
import numpy as np
import re
import plotly.express as px
import plotly.io as pio
from math import floor, ceil

pio.renderers.default = "notebook_connected"
```

```python
df = pd.read_json('runs.json')

descriptions = {
    'minCoinsToValue': 'Recursive',
    'minCoinsToValueCached': 'Recursive with Memoization'
}

df['description'] = df['algorithm'].map(descriptions)

list_cols = [col for col in df.columns if df[col].dtype == 'object' and df[col].apply(lambda x: isinstance(x, list)).all()]

df = df.drop(list_cols, axis=1)

df = df.drop(['meanPrimitiveCalls'], axis=1)

df
```

```python

fig = px.line(df, x="runArg", y="meanTime", color="description", title="Mean Elapsed Time", labels={"runArg": "Matrix Size", "meanTime": "Time (s)", "description": "Program Used"}, markers=True)

fig.show()
```

```python

# meanFunctionCalls

fig = px.line(df, x="runArg", y="meanFunctionCalls", color="description", title="Mean Function Calls", labels={"runArg": "Matrix Size", "meanFunctionCalls": "Function Calls", "description": "Program Used"}, markers=True)

fig.show()
```

```python
# scatter plot of function calls vs time

fig = px.scatter(df, x="meanFunctionCalls", y="meanTime", color="description", title="Function Calls vs Time", labels={"meanFunctionCalls": "Function Calls", "meanTime": "Time (s)", "description": "Program Used"}, log_x=True)

fig.show()
```
